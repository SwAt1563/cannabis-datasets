```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains

import time
import json
```

```python
driver = webdriver.Chrome()
driver.get("https://www.cannabisbusinesstimes.com/section/top-stories/")
time.sleep(2)
actions = ActionChains(driver)
```

```python
def load_more(driver):
    return driver.find_element(By.CSS_SELECTOR, "button.load-more")
```

```python
# remove ads
ads_remove_button = driver.find_element(By.CSS_SELECTOR, "i.CoverPop-close")
ads_remove_button.click()
```

```python
# Scroll to the 'load more' button
actions.move_to_element(load_more(driver)).perform()
```

```python
# load button
def load_all_meetings(driver):
    count = 0
    while True:
        try:
            load_button = load_more(driver)
            load_button.click()
            time.sleep(2)

            actions.move_to_element(load_more(driver)).perform()

            count += 1
            print(f'Loaded {count} times')
            if count > 20: # 20 is worst case scenario
                break
        except Exception as e:
            print(e)
            break
```

```python
load_all_meetings(driver)
```

```python
urls = []
```

```python
content = driver.find_element(By.CSS_SELECTOR, "div.content-list-container")
cards = content.find_elements(By.CSS_SELECTOR, "div.card")

for i in range(0, len(cards)):
    print(f'Processing card {i + 1}/{len(cards)}')
    card = cards[i]
    try:
        card_title = card.find_element(By.CSS_SELECTOR, "h3")
        card_link = card_title.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
        urls.append(card_link)
    except Exception as e:
        continue
```

```python
print(f'Found {len(urls)} news articles')
```

```python
# Save URLs to a JSON file
with open('urls.json', 'w') as file:
    json.dump(urls, file, indent=4)
```

```python
driver.quit()
```

```python
# Load URLs from the JSON file
with open('urls.json', 'r') as file:
    urls = json.load(file)
```

```python
driver = webdriver.Chrome()
```

```python
def get_page_data(url, driver):
    driver.get(url)

    # Wait to load page: not necessary
    # time.sleep(2)

    # remove ads
    try:
        ads_remove_button = driver.find_element(By.CSS_SELECTOR, "i.CoverPop-close")
        ads_remove_button.click()
    except Exception as e:
        pass

    
    try:
    
        main = driver.find_element(By.TAG_NAME, "main")
        title_element = main.find_element(By.TAG_NAME, "h1")
        title = title_element.text
        content = title_element.find_element(By.XPATH, "./following-sibling::p").text

        article = main.find_element(By.TAG_NAME, "article")
        line_text = article.find_element(By.CSS_SELECTOR, "div.byline").text
        author, date = line_text.split(" | ")

    except Exception as e:
        return None
    
    return {
        "title": title,
        "content": content,
        "author": author,
        "date": date,
        "url": url
    }
```

```python
news = []
```

```python
for i in range(0, len(urls)):
    print(f'Processing URL {i + 1}/{len(urls)}')
    url = urls[i]
    data = get_page_data(url, driver)
    if data:
        news.append(data)
    else:
        print(f'Failed to process URL {url}')
```

```python
# Save the results to a JSON file
with open('news.json', 'w') as file:
    json.dump(news, file, indent=4)
```

```python
# Cleanup
driver.quit()
```
